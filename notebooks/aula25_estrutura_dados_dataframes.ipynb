## Acessando dados especificos

### iloc - index location  
Localizar a informação atraves do index - tanto as linhas quanto as colunas devem ser passad o indice que indica a posição da linha e da coluna
- permite identificar apenas uma informação
iloc[INDEX DA LINHA, INDEX DA COLUNA]

- permite identificar mais informações
iloc[ INICO LINHA INDEX: FIM LINHA INDEX, INICIO COLUNA INDEX : FIM COLUNA INDEX ]

### loc - location - nome da coluna
 - tenho a linha como index e a coluna como nome

```python
df_selecionado = df.iloc[0, 0]
print(df_selecionado)
```

```python
#seleciona a primeira linnha e um intervalo de coluna
df_selecionado = df.iloc[0, 2:4]
print(df_selecionado)
```

```python
#seleciona um intervalo de linha e uma coluna
df_selecionado = df.iloc[1:5, 3]
print(df_selecionado)
```

```python
#seleciona um intervalo de linha e um intervalo de coluna
df_selecionado = df.iloc[0:3, 1:5]
print(df_selecionado)
```

```python
#seleciona um intervalo de linha e uma coluna pelo nome
df_selecionado = df.loc[0:4, "idade"]
print(df_selecionado)
```

```python
#seleciona um intervalo de linha e várias colunas pelo nome
colunas = ["idade", "estado", "cidade"]
df_selecionado = df.loc[:, colunas]
print(df_selecionado)
```

```python
#seleciona todas as linhas e todas as colunas pelo nome
df_selecionado = df.loc[:, :]
print(df_selecionado)
```

```python
#seleciona todas as linhas e todas as colunas pelo indice
df_selecionado = df.iloc[:, :]
print(df_selecionado)
```
